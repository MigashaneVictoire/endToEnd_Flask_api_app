```python
import pandas as pd
```

```python
df = pd.read_csv("./data/students.csv")
df.head()
```

```python
df["email"] = (df.first_name + df.last_name + "@gmail.com").str.strip().str.lower()

df.head()
```

```python
initials = []
for i in df.first_name.values:
    initials.append(i[0])

username = pd.Series(initials) + df.last_name
df["username"] = username
df
```

```python
df["password"] = (df.last_name + "pass" ).str.strip()

df.head()
```

```python
cred = df[["student_id", "email", "password"]].set_index('student_id')
cred.head()
```

```python
df.to_csv("./data/students.csv", mode="w")
```
